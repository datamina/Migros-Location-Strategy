```python
def migros_proportion_by_all_stores(list_of_cantons):
    all_store_count = {}
    migros_store_count = {}
    migros_proportion = {}

    for canton in list_of_cantons:
        canton_count = all_stores_df_with_cantons[all_stores_df_with_cantons['Canton_name'] == canton].shape[0]
        all_store_count[canton] = canton_count

        migros_count = all_stores_df_with_cantons[(all_stores_df_with_cantons['Canton_name'] == canton) & (all_stores_df_with_cantons['name'] == 'Migros')].shape[0]
        migros_store_count[canton] = migros_count

        if canton_count > 0:
            migros_proportion[canton] = migros_count / canton_count * 100  
        else:
            migros_proportion[canton] = 0    
            
        #print(f"In {canton}, there are {canton_count} supermarkets, out of which {migros_count} are Migros stores.\nSo, {migros_proportion[canton]:.2f}% of shops are Migros.")

    df = pd.DataFrame({'Canton': list_of_cantons,
        'Canton_count': [all_store_count[c] for c in list_of_cantons],
        'Migros_count': [migros_store_count[c] for c in list_of_cantons],
        'Proportion_Migros': [migros_proportion[c] for c in list_of_cantons]
    })
    
    #df.to_csv('migros_proportion_by_all_stores.csv', index=False)  
    
    return migros_store_count

migros_proportion_by_all_stores(list_of_cantons)
```

## Get proportion of Migros per population

```python
def migros_proportion_by_canton_population(list_of_cantons):
    migros_store_count = migros_proportion_by_all_stores(list_of_cantons)
    migros_ratio = {}

    population_by_canton_count = {
        canton: cantons_pop_df.loc[cantons_pop_df['Canton_name'] == canton, 'Population'].values[0] for canton in list_of_cantons}

    for canton in list_of_cantons:
        # Count the number of Migros stores in the given canton
        migros_count = all_stores_df_with_cantons[
            (all_stores_df_with_cantons['Canton_name'] == canton) & 
            (all_stores_df_with_cantons['name'] == 'Migros')
        ].shape[0]

        if migros_count > 0:
            ratio = population_by_canton_count[canton] / migros_count
        else:
            ratio = None
        
        migros_ratio[canton] = ratio
        #sorted_migros_proportion = dict(sorted(migros_proportion.items(), key=lambda item: item[1], reverse=True))  # find another way because None value

    df = pd.DataFrame({'Canton': list_of_cantons,
            'Population_count': [population_by_canton_count[c] for c in list_of_cantons],
            'Migros_count': [migros_store_count[c] for c in list_of_cantons],
            'Potential_customer by shop': [migros_ratio[c] for c in list_of_cantons]
        })

        #print(f"In {canton}, there are {population_by_canton_count[canton]} inhabitants and {migros_count} Migros stores.\n"
              #f"So, 1 shop for {ratio:.2f} people." if ratio else "No Migros stores in this canton.")

    #df.to_csv('migros_proportion_by_population.csv', index=False)  

    return df

migros_proportion_by_canton_population(list_of_cantons)
```
